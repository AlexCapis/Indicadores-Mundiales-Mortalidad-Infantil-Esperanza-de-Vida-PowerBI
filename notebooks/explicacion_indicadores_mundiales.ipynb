# Análisis de Población, Mortalidad Infantil y Esperanza de Vida en Power BI

## Introducción

Este proyecto se compone de dos partes donde en un primer lugar, tiene como objetivo realizar un análisis detallado de la población por área distinguiendo a los continentes de África, América, Asia, Europa y Oceania junto con sus respectivos países utilizando Power BI. En segundo lugar, también se analiza la mortalidad infantil y la esperanza de vida en estas mismas regiones.

A lo largo de este notebook, explicaré las transformaciones de datos, visualizaciones y segmentaciones que he aplicado al conjunto de datos original para proporcionar una visión clara y detallada.

**Requisitos Previos:**

Para ejecutar este proyecto, asegúrate de tener Power BI instalado y los conjuntos de datos `Countries.xlsx`, `Infant+death+rate.xlsx`, `Life+expectancy.xlsx`, `Paises.xlsx`, `Population.xlsx`, disponible en la carpeta "[data](https://github.com/AlexCapis/Ventas-PowerBI/tree/main/data)" del repositorio.

## Configuración de los Datos

En esta fase se ha realizado un profundo desarrollo en cuanto a las transformaciones del conjunto de datos a través del editor de Power Query para así obtener los siguientes datos:

1. Del archivo `Population.xlsx` se ha obtenido una columna con la cantidad de población ordenada por cantidades, las cuales he establecido de la siguiente manera:
- 0 - 1 Millón
- 1 M - 10 Millones
- 10 M - 100 Millones
- 100 Millones o más
Los he establecido de esta manera para así luego en Power Bi generar una segmentación de datos y que sea más sencillo a la hora de poder visualizar los datos de forma más dinámica.

2. Del archivo `Life+expectancy.xlsx` se ha obtenido una columna con la cantidad de esperanza de vida de la población ordenada por cantidades, las cuales he establecido de la siguiente manera:
- 0 a 60 años de edad
- 60 a 70 años de edad
- 70 a 80 años de edad
- 80 o más años de edad

3. Del archivo `Infant+death+rate.xlsx` se ha obtenido una columna con la cantidad de mortalidad infantil (muertes/1000 niños) de la población ordenada por cantidades, las cuales he establecido de la siguiente manera:
- 0 a 10 defunciones infantiles
- 10 a 25 defunciones infantiles
- 25 a 50 defunciones infantiles
- 50 o más defunciones infantiles

### Indicadores Mundiales

A través de esta página del dashboard he querido plasmar un análisis de mortalidad infantil y esperanza de vida por regiones. Se ha podido realizar con la ayuda de diversos tipos de visualizaciones como dos matrices, un gráfico de dispersión, un gráfico de barras agrupadas, un mapa, diversas segmentaciones de datos 2 botones (atrás y restablecer)  

![Mortalidad Infantil y la Esperanza de Vida por Regiones](./docs/images/mortalidad_infantil_esperanza_vida.png)

#### 1. Matriz

##### Matriz de población, esperanza de vida y mortalidad infantil por continente y país

A través de la primera matriz he querido plasmar las columnas de continente que se puede desglosar por países, la columna de población para ver la cantidad de esa que se ve reflejada por el color verde estableciendo que a mayor intensidad del color mayor cantidad de población, la columna de la esperanza de vida donde a través del color rojo he querido establecer que a menor esperanza de vida el color rojo será más intenso y por último, la columna de mortalidad infntil donde también he usado el color rojo para establecer que a mayor intensidad de color rojo será debido a que hay una mayor mortalidad infantil. 

Esta matriz resulta muy completa y al añadir los colores se puede visualizar de forma más rápida donde están los factores que más nos interesa saber en cuanto al continente o al país en base a la pblación, la esperna de vida y la mortalidad infantil.

![Matriz de población, esperanza de vida y mortalidad infantil por continente y país](../docs/images/matriz_esperanza_vida_mortalidad_infantil.png)

##### Matriz de cantidad de población y la esperanza de vida

En cuanto a la segunda matriz que he realizado he tratado de reflejar la cantidad de pablación de forma ordenada y la esperanza de vida agrapada por el color rojo que variará de color a un rojo más intenso cuanto menor sea la esperanza de vida en base a la cantidad de población.

![Matriz de cantidad de población y la esperanza de vida](../docs/images/matriz_esperanza_vida.png)

#### 2. Gráfico de Dispersión

En dicho gráfico de dispersión he tratado de plasmar en el eje X la mortalidad infantil y en el eje Y la esperanza de vida. Todo ello lo he querido reflejar por continente y país, con lo cual cada punto si nos situamos encima de él nos aparecerá una pestaña indicandonos; el país, el continente, la mortalidad infantil y la esperanza de vida. 

De esta manera podemos ver de forma más visual que países se encuentran más en zonas extremas como por ejemplo el caso de Angola (áfrica) y de Afganistán (Asia) tiene una gran mortalidad infantil y una esperanza de vida bastante baja, mientras que en el lado opuesto podemos observar a países como Mónaco (Europa) o Singapour (Asia), los cuales tienen una esperanza de vida muy elevada y una mortalidad infantil en niveles muy bajos. Esta manera de visualizar los datos es muy relevante y visual, ya que almacena una gran cantidad de datos con un simple vistazo.
